```python
import networkx as nx
import deepwalk
```

```python
path = "drive/MyDrive/ColabNotebooks/Decagon_Embedding/"
G = nx.read_edgelist(path+"PP-Decagon_ppi.csv", delimiter=",")
print(len(G.nodes), len(G.edges))
n_shuffles = [5, 10, 15, 25, 50]
Gs_shuffled = dict()
```

```python
for n_shuffle in n_shuffles:
  Gs_shuffled["G_{}".format(n_shuffle)] =  nx.read_edgelist(path+"PP-Decagon_ppi_{}.csv".format(n_shuffle), delimiter=" ") #delimiter=","
Gs_shuffled
```

```python
dimensions = int(input("Enter Embedding dimensions: "))
walk_length = int(input("Enter the lenght of each random walk: "))
num_walks = int(input("Enter the number of walks per node: "))

for n_shuffle in n_shuffles:
    
    print("n_shuffles = {}".format(n_shuffle))
    input_file = path+"PP-Decagon_ppi_{}.csv".format(n_shuffle)
    output_file = path+"Outputs/"+"G_{}_DeepWalkEmbedding".format(n_shuffle)
    !deepwalk --input  $input_file --format edgelist --number-walks  $num_walks --output  $output_file --representation-size  $dimensions --walk-length  $walk_length
```
