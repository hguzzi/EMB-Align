```python
dimensions = int(input("Enter Embedding dimensions: "))
order = int(input("Enter the order of the proximity to use (1 or 2): "))
samples = int(input("Enter the total number of training samples (*Million): "))

if order not in [1,2]:
  order = 2

for n_shuffle in n_shuffles:

    print("n_shuffles = {}".format(n_shuffle))
    output_file = path+"Outputs/"+"G_{}_LINE_Embedding".format(n_shuffle)
    input_file = path+"PP-Decagon_ppi_{}.csv".format(n_shuffle)
    !./line.exe -train $input_file -output $output_file -size $dimensions -order $order -samples $samples
```
